```python
from pandas.core import frame
from scipy.sparse import data

#CHAPTER-1, pages from 14 to 23

from sklearn.datasets import load_iris
iris_dataset = load_iris()  #load_iris() returns object "iris" which is very similar to dictionary (contains keys and values)

print ("Keys of dataset IRIS: \n{}".format(iris_dataset.keys()))  # ????THESE TWO ARE THE SAME. THEN WHY USE THE FIRST LONG VERSION??????????
print (iris_dataset.keys())                                       # ????THESE TWO ARE THE SAME. THEN WHY USE THE FIRST LONG VERSION??????????

print("\n\n\n\n\nVALUE OF THE KEY -->DESCR<--")
print(iris_dataset['DESCR'][:1100] + "\n...")     #we requested 1100 characters of DESC string.. Update 1100 and see what changes in output.

print("\n\n\n\n\nVALUE OF THE KEY -->Target Names<--")
print("Target Names: {}".format(iris_dataset['target_names']))

print("\n\n\n\n\nVALUE OF THE KEY -->data<-- - first 7 rows:\n")
print(format(iris_dataset['data'][:7]))

print("\n\n\n\n\nVALUE OF THE KEY  -->target<--\n")
print("Target:\n{}".format(iris_dataset['target']))
print("\nTARGET is a one dimensional array, with one entry per flower.\n")
print("Let's now once more extract the meanings of the numbers in TARGET:\n")
print(format(iris_dataset['target_names']))

print("\n\n\n\n\nVALUE OF THE KEY  -->frame<--\n")
print(format(iris_dataset['frame']))


print("\n\n\n\n\nVALUE OF THE KEY  -->feature_names<--")
print(format(iris_dataset['feature_names']))

print("\n\n\n\n\nVALUE OF THE KEY   -->filename<--")
print(format(iris_dataset['filename']))


print("\n\n\n\n\nVALUE OF THE KEY   -->data_module<--")
print(format(iris_dataset['data_module']))

print("\n\n\n\nALL THE DATA IN IRIS DATASET:\n")
print(format(iris_dataset.data)) #SELF TRY

print("\n\n\n\n")
print("SHAPE OF DATA: {}".format(iris_dataset['data'].shape))  #outputs (150,4) that means: 150 samples each with 4 properties 



# LET'S SPLIT THE WHOLE IRIS DATASET TO TRAIN-SET AND TEST-SET
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( iris_dataset['data'],  iris_dataset['target'], random_state=0)
print("\n\n\nWe have splitted the whole IRIS dataset to TRAIN SET and TEST SET. After splitting we have:")
print("X_train shape:{}".format(X_train.shape))
print("y_train shape:{}".format(y_train.shape))
print("X_test shape:{}".format(X_test.shape))
print("y_test shape:{}".format(y_test.shape))



# ONE MORE STEP BEFORE BUILDING OUR MODEL: LOOKING AT OUR DATA.
# WE WILL GENERATE SOME PAIR-PLOTS OF IRIS DATASET USING scatter_matrix FUNCTION IN PANDAS Library
# create dataframe from data in X_train
import pandas as pd
#print(pd.__version__)
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names) #label the columns using the strings in iris_dataset.feature_names:
print("create a scatter matrix from the dataframe, color by y_train\n")
grr=pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(15,15), marker='*', hist_kwds={'bins':20}, s=60, alpha=.8)


print("WE DECIDED TO USE K-NEAREST NEIGHNORS CLASSIFICATION ALGORITHM...")
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)   #number of neighbors is the most important paramater which will be used by knn to decide on classification
print(knn.fit(X_train,y_train))
print("\nWe have found a new iris flower with a sepal lenght of:5cm, a sepal width of:2.9cm \n a petal length of 1cm and a petal width of 0.2cm. \n What species of iris would this be?")

import numpy as np
X_new=np.array([[5, 2.9, 1, 0.2]])
print("X_new.shape: {}".format(X_new.shape))
#Now let's make the prediction:
prediction=knn.predict(X_new)
print("Prediction: {}".format(prediction))
print("Predicted target name: {}".format(iris_dataset['target_names'][prediction]))

print("CAN WE TRUST THE CLASSIFICATION RESULT?")
#below we make some accuracy tests to evaluate the prediction of our model:
y_pred=knn.predict(X_test)
print("Test set predictions:\n {}".format(y_pred)) 
print("Test score: {:.2f}".format(np.mean(y_pred==y_test)))
#Another more practical way to compute test score is: the SCORE method of knn object:
print("Test set score: {:.2f}".format(knn.score(X_test,y_test)))











print("\n\n\n\n\n\n\n\n")
#BELOW CODE BLOCK GOT FROM THE WEB SITE:
#https://www.geeksforgeeks.org/plotting-graph-for-iris-dataset-using-seaborn-and-matplotlib/
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
data = pd.read_csv("Iris.csv")
   
print (data.head(10))
"""
#IT GIVES ERROR!!???????????????????????????????????????



```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets
from sklearn.inspection import DecisionBoundaryDisplay

n_neighbors = 15

# import some data to play with
iris = datasets.load_iris()

# we only take the first two features. We could avoid this ugly
# slicing by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

# Create color maps
cmap_light = ListedColormap(["orange", "cyan", "cornflowerblue"])
cmap_bold = ["darkorange", "c", "darkblue"]

for weights in ["uniform", "distance"]:
    # we create an instance of Neighbours Classifier and fit the data.
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X, y)

    _, ax = plt.subplots()
    DecisionBoundaryDisplay.from_estimator(
        clf,
        X,
        cmap=cmap_light,
        ax=ax,
        response_method="predict",
        plot_method="pcolormesh",
        xlabel=iris.feature_names[0],
        ylabel=iris.feature_names[1],
        shading="auto",
    )

    # Plot also the training points
    sns.scatterplot(
        x=X[:, 0],
        y=X[:, 1],
        hue=iris.target_names[y],
        palette=cmap_bold,
        alpha=1.0,
        edgecolor="black",
    )
    plt.title(
        "3-Class classification (k = %i, weights = '%s')" % (n_neighbors, weights)
    )

plt.show()
```
